## Testes mensurador

### Exemplo de texto 1

Medindo documento com sentenças.

```
# Documento e suas sentenças
DO = ["Bom Dia, professor.",
      "Qual o conteúdo da prova?",
      "Vai cair tudo na prova?",
      "Aguardo uma resposta, João."]

# Recupera a medida Ccos do documento
saida = modelo.getMedidasTexto(DO, palavra_relevante=0)

# Mostra a medida recuperada
print("Ccos DO    :", saida['cos']) #Ccos DO    : 0.8178287347157797
print("Ceuc DO    :", saida['euc']) #Ccos DO    : 0.8178287347157797
print("Cman DO    :", saida['man']) #Ccos DO    : 0.8178287347157797
```

```
# Documento e suas sentenças
DO = ["Bom Dia, professor.",
      "Qual o conteúdo da prova?",
      "Vai cair tudo na prova?",
      "Aguardo uma resposta, João."]

# Recupera a medida Ccos do documento
CcosDO = modelo.getMedidasTextoCosseno(DO, palavra_relevante=0)

# Mostra a medida recuperada
print("Ccos DO    :", CcosDO) #Ccos DO    : 0.8178287347157797
```

### Exemplo de texto 2

Comparando documento original(DO) com documento com sentenças permutadas(permDO).

```
# Define documento original com 4 sentenças
DO1 = ["Bom Dia, professor.",
      "Qual o conteúdo da prova?",
      "Vai cair tudo na prova?",
      "Aguardo uma resposta, João."]

# Define um documento com a permutação das sentenças do documento original
permDO1 = ["Aguardo uma resposta, João.",
           "Qual o conteúdo da prova?",
           "Bom Dia, professor.",
           "Vai cair tudo na prova?"]

# Recupera a medida Ccos dos documentos
CcosDO1 =  modelo.getMedidasTextoCosseno(DO1)
CcospermDO1 =  modelo.getMedidasTextoCosseno(permDO1)

# Mostra as medidas recuperadas
print("Ccos DO    :", CcosDO1)
print("Ccos permDO:", CcospermDO1)
```
